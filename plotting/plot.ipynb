```python


import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('architecture_results (4).csv')
df
```

```python
df.columns
```

```python
df2 = df[df['network_recovery_rate'] == 10]
```

```python
df1 = df2[['Seed', 'Average routers per node', 'network_recovery_rate', 'Routers', 'Statistics.online_tx', 'Statistics.offline_tx']]
df1
```

```python
df1['Offline rate'] = df1['Statistics.online_tx'] / (df1['Statistics.online_tx'] + df1['Statistics.offline_tx'])
#df1['Avarage routers per node'] = str(df['Avarage routers per node'])
df1
```

```python
df3 = df1.copy()
#
# df3['Routers'] = df3['Routers'].apply(lambda x: np.log(x))
#df3 = df3[df3['Routers'] >= 5]
df3
```

```python
sns.set_theme(style="darkgrid")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(12, 10))

# Plot the orbital period with horizontal boxes
sns.boxplot(x="Average routers per node", y="Offline rate", data=df3,
            whis=[0, 100], width=.6, palette="vlag")

# Add in points to show each observation
#sns.stripplot(x="Average routers per node", y="Offline rate", data=df3,
#              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
plt.ylim(0)
ax.xaxis.grid(True)
ax.set(ylabel="Rate of online transactions", xlabel="Intermediaries per client")
sns.despine(trim=True, left=True)
```

```python


sns.set_theme(style="dark")
# Plot each year's time series in its own facet
g = sns.relplot(
    data=df3,
    x="Average routers per node", y="Offline rate",
    kind="line", palette="crest", linewidth=4, zorder=5, height=4, aspect=1.5, legend=False,
)

g.set_axis_labels("Average intermediaries per node", "Online tx rate")
g.tight_layout()
```

```python
b = sns.lineplot(data=df3, x="Routers", y="Offline rate", hue="Average routers per node", style="Average routers per node", legend="auto")
#b.set_axis_labels("Intermediaries", "Online tx rate")
b.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#b.tight_layout()
```
